```python
arr_abs=[]
for i in arr:
    if i<0:
        arr_abs.append(-i)
    else :
        arr_abs.append(i)
arr_abs
```

```python
arr_abs=[-i if i<0 else i for i in arr] # arr에서 i로 하나씩 가져와서 if만약 i가 0보다작으면 왼쪽, 그렇지않으면 오른쪽실행한다
arr_abs
```

$$ \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{h} $$

```python
import os
```

```python
def derivative(f,x):
    h=0.000001
    return [(f(i+h)-f(i))/h for i in x]
```

```python
fs=[np.cos,np.sin,np.exp,np.log,lambda i:i*2+3, lambda i:i**2]
fss=["cos","sin","exp","log","2x^3+3","x^2"]
```

```python
fn=[np.cos, np.sin, np.exp, lambda i:np.log(abs(i)), lambda i: (i**3)*2+3, lambda i:i**2]
x=np.arange(-10,10,0.01)
result=[[f(i)for i in x] for f in fn]
```

```python
for i,f in enumerate(zip((fn),fss)):
    plt.subplot(6,2,i*2+1)
    plt.title(f'{f[1]} funtion')
    plt.ylabel(f'{f[1]} value')
    plt.plot(x, result[i])
    plt.subplot(6,2,i*2+2)
    plt.title(f'{f[1]} derivative funtion')
    plt.ylabel(f'{f[1]} value')
    plt.plot(x, derivative(f[0],x))
```
