```python
import cv2
import numpy as np
```

```python
# 승주가 짠 코드 (마우스클릭시 화면이 내려가서 눈이내리는것 같음)
img=np.full((800,1200,3),50,np.uint8)
cnt=0
s_dict={}
def on_mouse(event,x,y,flags,params):
    global  cnt ,s_dict
    if event==cv2.EVENT_LBUTTONDOWN:
        s_dict[cnt]=[x,y]
        cnt+=1
    for i in s_dict.values():
        i[1]+=1
    
    img=np.full((800,1200,3),50,np.uint8)
    del_list=[]
    for i,j in s_dict.items():
        if j[1]>1300:
            del_list.append(i)
        cv2.drawMarker(img,(j[0],j[1]),(255,255,255),cv2.MARKER_STAR,10)
        cv2.imshow('img',img)
    for i in del_list:
        del s_dict[i]
cv2.namedWindow('img')
cv2.setMouseCallback('img',on_mouse)
cv2.imshow('img',img)
cv2.waitKey()
cv2.destroyAllWindows
```

```python
import cv2
import numpy as np

img = np.full((800, 1200, 3), 50, np.uint8)

for t in range(10):
    # 랜덤한 위치 선택
    x = np.random.randint(0, 1200)
    y = np.random.randint(0, 800)
    color_ = (np.random.randint(150,255),np.random.randint(150,255),np.random.randint(150,255))
    # 별 모양 좌표 생성
    star_points = np.array([
        [x, y - 30], [x - 10, y - 10],
        [x - 30, y - 10], [x - 15, y + 10],
        [x - 20, y + 30], [x, y + 20],
        [x + 20, y + 30], [x + 15, y + 10],
        [x + 30, y - 10], [x + 10, y - 10]
    ], dtype=np.int32)

    # 별 그리기
    cv2.polylines(img, [star_points], isClosed=True, color=(color_), thickness=2)

cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = np.full((800, 1200, 3), (130,220,130), np.uint8)

for t in range(20):
    # 랜덤한 위치 선택
    x = np.random.randint(0, 1200)
    y = np.random.randint(0, 800)
    color_ = (np.random.randint(150,255),np.random.randint(150,255),np.random.randint(150,255))
    # 랜덤한 마커 유형 선택
    random_marker = np.random.randint(0, 7)  # 0부터 6까지의 정수
    marker_types = [cv2.MARKER_CROSS, cv2.MARKER_TILTED_CROSS, cv2.MARKER_STAR,
                    cv2.MARKER_DIAMOND, cv2.MARKER_SQUARE, cv2.MARKER_TRIANGLE_UP,
                    cv2.MARKER_TRIANGLE_DOWN]
    
    # 선택한 마커 유형으로 마커 그리기
    cv2.drawMarker(img, (x, y), (color_), markerType=marker_types[random_marker], thickness=2)
    cv2.imshow('img', img)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import time

img = np.full((800, 1200, 3), 50, np.uint8)
marker_list = []  # 마커 좌표를 저장하는 리스트

while True:
    # 랜덤한 위치 선택
    x = np.random.randint(0, 1200)
    y = np.random.randint(0, 800)

    # 랜덤한 마커 유형 선택
    random_marker = np.random.randint(0, 7)  # 0부터 6까지의 정수

    # 선택한 마커 유형으로 마커 그리기
    cv2.drawMarker(img, (x, y), (255, 255, 255), cv2.MARKER_STAR, thickness=2)

    # 마커 좌표를 리스트에 추가
    marker_list.append((x, y))

    # 이전 마커들을 함께 그리기
    for marker in marker_list:
        cv2.drawMarker(img, marker, (255, 255, 255), cv2.MARKER_STAR, thickness=2)

    cv2.imshow('img', img)
    
    y -= 10
    # 2초 대기
    key = cv2.waitKey(1000)

    # 'q' 키를 누르면 루프 종료
    if key == ord('q'):
        break

cv2.destroyAllWindows()
```
