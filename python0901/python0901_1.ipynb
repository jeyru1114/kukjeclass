```python
import requests
from bs4 import BeautifulSoup
import urllib
```

```python
load_url = "http://python.cyber.co.kr/pds/books/python2nd/test2.html"
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
# 모든 a 태그를 검색하고 링크를 절대 URL로 표시
for i in soup.find_all("a"):
    print(i.text)
    url = i.get("href")
    print("url:",url)
    link_url = urllib.parse.urljoin(load_url, url)
    print(link_url)
```

```python
from html.parser import HTMLParser
```

```python
class StrongMyHTMLParser(HTMLParser): # HTMLParser 를 상속받아 클래스 설계 (A)
    def __init__(self): #자식 생성자(B)
        HTMLParser.__init__(self) #부모 생성자(A)
        self.is_strong=False
    def handle_starttag(self, tag, attrs):
        if tag == 'strong' : #<strong>태그 시작
            self.is_strong=True
    def handle_endtag(self, tag): # 부모 클래스 오버라이딩, HTMLParser안에 handle_endtag가 있음
        if tag=='strong' : #<strong>태그 닫힘
            self.is_strong=False
    def handle_data(self, data):
        if self.is_strong: #<strong> ~~~~ </strong> 구간인 경우
            print(data) # 데이터 출력
```

```python
with open('11.html') as f:
    parser = StrongMyHTMLParser()
    parser.feed(f.read())
```

```python
class LIMyHTMLParser(HTMLParser): # HTMLParser 를 상속받아 클래스 설계 (A)
    def __init__(self): #자식 생성자(B)
        HTMLParser.__init__(self) #부모 생성자(A)
        self.is_strong=False
    def handle_starttag(self, tag, attrs):
        if tag == 'li' : #<strong>태그 시작
            self.is_strong=True
    def handle_endtag(self, tag): # 부모 클래스 오버라이딩, HTMLParser안에 handle_endtag가 있음
        if tag=='li' : #<strong>태그 닫힘
            self.is_strong=False
    def handle_data(self, data):
        if self.is_strong: #<strong> ~~~~ </strong> 구간인 경우
            print(data) # 데이터 출력
```

```python
with open('11.html') as f:
    parser = LIMyHTMLParser()
    parser.feed(f.read())
```

handle_starttag(tag, attrs): 시작 태그를 만났을 때 호출되며, 태그 이름과 속성의 리스트를 인자로 받습니다.

handle_endtag(tag): 종료 태그를 만났을 때 호출되며, 태그 이름을 인자로 받습니다.

handle_data(data): 태그 내의 텍스트 데이터를 만났을 때 호출되며, 데이터 내용을 인자로 받습니다.

handle_comment(data): 주석을 만났을 때 호출되며, 주석 내용을 인자로 받습니다.

handle_entityref(name): HTML 엔티티 레퍼런스(예: &amp;, &lt;)를 만났을 때 호출되며, 엔티티 이름을 인자로 받습니다.

handle_charref(name): 문자 참조(예: &#123;)를 만났을 때 호출되며, 참조한 숫자를 인자로 받습니다.

handle_decl(decl): 선언 (예: <!DOCTYPE>)을 만났을 때 호출되며, 선언 내용을 인자로 받습니다.

```python
load_url = "http://python.cyber.co.kr/pds/books/python2nd/test2.html"
```

```python
html = requests.get(load_url)
soup = BeautifulSoup(html.content, "html.parser")
```

```python
filename = "linklist.txt"
with open(filename, "w") as f:
    for i in soup.find_all("a"):
        url = i.get("href")
        link_url = urllib.parse.urljoin(load_url, url)
        f.write(i.text+"\n")
        f.write(link_url+"\n")
        f.write("\n")
```
