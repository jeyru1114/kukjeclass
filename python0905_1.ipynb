```python
def deg_to_rad(x): #각도 => 라디안 (pi = 3.14)
    return (x*np.pi)/180
```

```python
def rad_to_deg(x): # 라디안(pi = 3.14) => 각도 (180)
    return (x/np.pi)*180
```
