````python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy import symbols
````

````python
x1, y1 = symbols('x y') # sympy data

def f1():
    x = np.arange(-10,10,0.01)
    y = x**2+3*x+2
    y1 = x1**2+3*x1+2
    y_derivative = np.diff(y)
    y_sym = diff(y1,x1) # sympy data
    print('y_sym:', y_sym)
    y_sym_after = [2*i + 3 for i in x] # # sympy data를 numpy data로 변환
    plt.plot(x,y)
    plt.plot(x, y_sym_after) # 미분하면 2x+3
    plt.show() # 추가
````

````python
window = tk.Tk()
button = tk.Button(text='버튼', width=50, height=10, bg='pink', fg='white', command=f1)
button.pack()
window.mainloop()
````

````python
def derivative(f, x, h=1e-5): # 미분의 정의
    return (f(x+h)-f(x))/h
````

````python
def origin_f(x_): # x^2 + 3x + 2
    return x_**2+3*x_+2
````

````python
def f_test():
    x_ = np.arange(-10,10,0.01)
    y = origin_f(x_)
    y_dev = derivative(origin_f, x_)
    plt.plot(x_, y)
    plt.plot(x_, y_dev)
    plt.show()
````

````python
window = tk.Tk()
button = tk.Button(text='버튼', width=50, height=10, bg='pink', fg='white', command=f_test)
button.pack()
window.mainloop()
````

````python
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo
from calendar import month_name, day_name
````

````python
root = tk.Tk()
````

````python
#root 윈도우 구성
root.geometry('300x200')
root.resizable(False,False)
root.title("콤보박스 widget")
````

````python
label = ttk.Label(text = '원하는 달 선택해주세요')
label.pack(fill = tk.X , padx=5, pady=5)
````

````python
#콤보 박스 생성
selected_month = tk.StringVar() # 위젯의 값을 효율적으로 관리하기 위한 객체
month_cb = ttk.Combobox(root, textvariable = selected_month)
#위젯의 textvariable에 selected_month 를 할당
#모든 달의 3개의 문자만 추출
month_cb['values'] = [month_name[m][:3] for m in range(1,13)]

#값 typing 하는것 방지
month_cb['state'] = 'readonly'
#위젯 배치
month_cb.pack(fill = tk.X , padx=5, pady=5)
#선택된 값의 변화를 binding
def month_changed(event):
    """
        month 가 변경된 event를 처리하는 함수(event handler, callback)
    """
    showinfo(title = 'Result', message = f'당신은{selected_month.get()}을 선택하셨어요')
month_cb.bind('<<ComboboxSelected>>', month_changed)
root.mainloop()
````
