## Sigmoid or Logistic Function
<img align="left" src="./images/C1_W3_LogisticRegression_left.png"     style=" width:300px; padding: 10px; " >classification task(분류 작업),linear regression model(선형 회귀) model을 사용하여 시작 $f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = \mathbf{w} \cdot  \mathbf{x}^{(i)} + b$, $y$ 주어진 값을 $x$ 예측. 
- 그러나, classification model의 예측값이  0 과 1 사이이다, 출력값이 $y$ 는 0 이나 1 중의 하나다 
- 모든 입력값은  0 과 1 사이로 mapping 하기 위한 용도로 "sigmoid function"를 사용한다. 

## Formula for Sigmoid function

sigmoid function 공식은 아래와 같다  

$g(z) = \frac{1}{1+e^{-z}}\tag{1}$

logistic regression 의 경우, linear regression model 의 결과가 z (sigmoid function의 입력으로 들어간다)<br>
$ z = w_{1}x_{1} + w_{2}x_{2} + \cdot\cdot\cdot + w_{n}x_{n} + b $
$ 이 z가 sigmoid 함수의 입력으로 들어간다 $
$ z = w^{T} x + b \rightarrow w, x, b는 벡터다(벡터 표현) $

```python
def sigmoid(z):
    """
        z의 sigmoid 함수 계산
        매개변수 :
            z(ndarray) : 하나의 스칼라, 임의의 크기를 가진 numpy배열
        반환값 :
        g(ndarray): sigmoid(z), 입력값(z) 와 같은 형태(shape)
        
    """
    g = 1/(1+np.exp(-z))
    return g
```

## Logistic Regression
<img align="left" src="./images/C1_W3_LogisticRegression_right.png"     style=" width:300px; padding: 10px; " >logistic regression model은 익숙한 linear regression model 에 아래에서 보는바와 같이 시그모이드(g, $\sigma$)를 적용함

$$ f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(\mathbf{w} \cdot \mathbf{x}^{(i)} + b ) \tag{2} = $$
$$ \sigma(\mathbf{w} \cdot \mathbf{x}^{(i)} + b ) \tag{2} $$ 

  여기에서

  $g(z) = \frac{1}{1+e^{-z}}\tag{3}$

## Squared error for logistic regression?
<img align="left" src="./images/C1_W3_SqErrorVsLogistic.png"     style=" width:400px; padding: 10px; " > Recall for **Linear** Regression we have used the **squared error cost function**:
The equation for the squared error cost with one variable is:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 \tag{1}$$ 
 
where 
  $$f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{2}$$
Now, let's get a surface plot of the cost using a *squared error cost*:
  $$J(w,b) = \frac{1}{2m} \sum\limits_{i = 0}^{m-1} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$ 
 
where 
  $$f_{w,b}(x^{(i)}) = sigmoid(wx^{(i)} + b )$$
